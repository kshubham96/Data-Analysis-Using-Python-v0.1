```python
#Chapter 8: Concept of Inner Join by creating two tables.
dataframe1 = pd.DataFrame({
    "employee": ["ABC", "XYZ", "MNO"],
    "age": ["22", "25", "30"]
})

dataframe1
```

```python
#Chapter 8: Concept of Joins by creating two tables.
dataframe2 = pd.DataFrame({
    "employee": ["ABC", "XYZ", "PQR"],
    "salary": ["10000", "12500", "30000"]
})

dataframe2
```

```python
dataframe3 = pd.merge(dataframe1,dataframe2, on="employee")
dataframe3
```

```python
#Outer
dataframe3 = pd.merge(dataframe1,dataframe2, on="employee", how="outer")
dataframe3
```

```python
#Left
dataframe3 = pd.merge(dataframe1,dataframe2, on="employee", how="left")
dataframe3
```

```python
#Right
dataframe3 = pd.merge(dataframe1,dataframe2, on="employee", how="right")
dataframe3
```

```python
fb_finance = pd.read_csv('C:/Users/dell/Desktop/shubham/Stuff/FB.csv',index_col = 'Date')
fb_finance.head()
```

```python
fb_finance.shift(-1).head()
```

```python
# Calculate the stock price delta
fb_finance['previous_closing_price'] = fb_finance['Close'].shift(1)
fb_finance.head()
```

```python
fb_finance['closing_price_delta'] = fb_finance['Close'] - fb_finance['previous_closing_price']
fb_finance.head()
```

```python
# Weekly returns
fb_finance['Weekly returns'] = ((fb_finance['Close'] - fb_finance['Close'].shift(7))/fb_finance['Close'])*100
fb_finance
```

```python
# Write your data to csv

fb_finance.to_csv('C:/Users/dell/Desktop/shubham/Stuff/FB.csv')
```
